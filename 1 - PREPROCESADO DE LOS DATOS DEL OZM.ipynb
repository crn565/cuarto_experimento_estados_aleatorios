# Creacion de los ficheros de datos en formato csv a partir de los ficheros en formato parquet

A  continuacion vamos a hacer una serie  de manipulaciones con los datos aportados en los ficheros de datos en formato novedoso comprimido parquet.

Resumidamente  estos son los pasos que vamos a seguir:


- En primer lugar cargaremos los ficheros en un dataset 
- En segundo lugar calcularemos la potencia aparente en funcion de dos columnas ( pot real y pot reactiva)
- En tercer lugar eliminaremos las columnas qeu no vamos a necesitar
- Seguidamente reordenamos las columnas en el orden que necesitamos
- Finalmente salvamos el dataset obtenido en un fichero en formato csv

## GeneraciÃ³n del fichero de datos a partir del fichero fan.csv  

```python
import pandas as pd
import numpy as np

with open(r'.\data\ozm1\ozm_fan_200MS.parquet', 'rb') as f:
    df_datos_agregados = pd.read_parquet(f, engine='pyarrow')
    

df_datos_agregados.rename(columns={'Time':'timestamp','Active_Power':'W','Reactive_Power':'VAR','Apparent_Power':'VA'} ,inplace=True)
df_datos_agregados.rename(columns={'Frequency':'f','Voltage':'VLN','Power_Factor':'PF','Current':'A' },inplace=True)



df_datos_agregados.drop (['FirstTimestamp','OriginalTimestamp','Tag'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Flag','Unbalance','Phi'],axis = 'columns', inplace=True)



df_datos_agregados=df_datos_agregados [['timestamp','W','VAR','VA','f','VLN','PF','A']]

print (df_datos_agregados)

df_datos_agregados.to_csv('fan.csv')


```
